```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
import pandas as pd
import os

# ==========================================
# 1. DATA LOADER (Based on Untitled3.ipynb)
# ==========================================
def load_trajectory_data(file_path):
    """
    Parses the trajectory text files provided in the project.
    Returns a dictionary of numpy arrays for time, x, h, etc.
    """
    data = []
    header = []

    if not os.path.exists(file_path):
        print(f"Warning: File '{file_path}' not found. Skipping data overlay.")
        return None

    try:
        with open(file_path, "r") as file:
            lines = file.readlines()
            start_data_index = -1
            # Find the first line that starts with a digit to identify the start of data
            for i, line in enumerate(lines):
                stripped_line = line.strip()
                if not stripped_line:
                    continue
                # Data lines are assumed to start with a number or a negative number
                if stripped_line[0].isdigit() or (stripped_line[0] == '-' and len(stripped_line) > 1 and stripped_line[1].isdigit()):
                    start_data_index = i
                    break

            if start_data_index == -1:
                print(f"Error: No data lines found in {file_path}")
                return None

            # The header is assumed to be the line immediately preceding the data
            if start_data_index > 0:
                header_line = lines[start_data_index - 1].strip()
                # Initial split and strip
                header = [h.strip() for h in header_line.split("|")]
                # Further cleaning: remove commas, strip leading/trailing underscores, make lowercase
                header = [h.replace(",", "").strip("_").lower() for h in header if h.strip()]
                # Filter out any empty strings that might result from extra delimiters or cleaning
                header = [h for h in header if h]
            else:
                print(f"Warning: Could not find header for {file_path}. Proceeding without header.")
                return None # Or handle more gracefully

            # Process data lines
            for line in lines[start_data_index:]:
                stripped_line = line.strip()
                if not stripped_line:
                    continue
                # Split by '|', filter out empty items, convert to float
                row = []
                for item in stripped_line.split("|"):
                    cleaned_item = item.strip()
                    if cleaned_item: # Only process non-empty strings
                        try:
                            row.append(float(cleaned_item))
                        except ValueError:
                            print(f"Warning: Could not convert '{cleaned_item}' to float in {file_path}. Skipping row or item.")
                            row = [] # Discard current row if any item fails conversion
                            break
                if row and len(row) == len(header): # Ensure row is not empty and matches header length
                    data.append(row)
                elif row:
                    print(f"Warning: Row length ({len(row)}) does not match header length ({len(header)}) for a data row in {file_path}. Skipping row.")


        if not data:
            print(f"No valid data found in {file_path}.")
            return None

        data_np = np.array(data)
        if len(header) != data_np.shape[1]:
            print(f"Error: Mismatch between number of headers ({len(header)}) and number of data columns ({data_np.shape[1]}) after parsing {file_path}.")
            print(f"Headers: {header}")
            return None

        dataset = {h: data_np[:, i] for i, h in enumerate(header)}
        return dataset

    except Exception as e:
        print(f"Error parsing {file_path}: {e}")
        return None

# ==========================================
# 2. AIRCRAFT PARAMETERS (B787-8)
# ==========================================
class B787:
    def __init__(self):
        # Mass and Geometry (Imperial Units to match ASE 367K context)
        self.W = 502500.0       # Max Takeoff Weight (lbs)
        self.S = 3501.0         # Wing Area (ft^2)
        self.b = 197.0          # Wingspan (ft)
        self.c_bar = 20.0       # Mean Aerodynamic Chord (ft)
        self.g = 32.174         # Gravity (ft/s^2)
        self.m = self.W / self.g # Mass (slugs)

        # Aerodynamics
        self.rho = 0.002377     # Air density at SL (slugs/ft^3)
        self.CD0_clean = 0.015  # Parasitic drag coefficient
        self.CD0_flaps = 0.020  # Drag penalty for takeoff flaps
        self.CD0_gear  = 0.015  # Drag penalty for landing gear
        self.K = 0.045          # Induced drag factor (1 / (pi * e * AR))

        self.CL_max_to = 1.8    # Max Lift Coeff (Takeoff Flaps)
        self.CL_max_clean = 1.4 # Max Lift Coeff (Clean)

        # Propulsion (GEnx-1B)
        self.Thrust_max = 140000.0 # Total Thrust (lbf) approx for 2 engines

# ==========================================
# 3. SIMULATION DYNAMICS
# ==========================================
def flight_dynamics(t, state, ac, scenario):
    """
    3-DOF Point Mass Equations of Motion
    State: [V (ft/s), gamma (rad), h (ft), x (ft)]
    """
    V, gamma, h, x = state

    # --- Pilot / Autopilot Logic ---
    # Rotation Logic: Rotate at Vr
    Vr = 155.0 * 1.688 # Knots to ft/s
    if V < Vr:
        # On Ground
        alpha = 0.0 # Rolling on wheels
        on_ground = True
    else:
        # Rotate
        target_alpha = np.deg2rad(10.0) # Target climb AoA
        # Simple lag for rotation dynamics (approximate pitch response)
        alpha = min(target_alpha, np.deg2rad(10.0) * (V - Vr) / 20.0)
        on_ground = False

    if h > 5000: # Stop simulation if too high
        return [0,0,0,0]

    # --- Scenario Failure Logic ---
    # Trigger failures based on altitude or time (Positive Rate ~ h > 50ft)
    failure_trigger_h = 50.0

    # 1. Gear Status
    gear_down = True # Gear stays down in failure scenarios (accident report)
    if scenario == 'nominal' and h > 200:
        gear_down = False

    # 2. Flap Status
    flaps_down = True
    if scenario == 'flaps' and h > failure_trigger_h:
        flaps_down = False # Flaps retracted mistakenly

    # 3. Engine Status
    thrust_level = 1.0
    if scenario == 'engine' and h > failure_trigger_h:
        thrust_level = 0.0 # Fuel cutoff

    # --- Aerodynamic Coefficients ---
    # CL Calculation
    CL0 = 0.2 if not flaps_down else 0.6
    CL_alpha = 5.5
    CL = CL0 + CL_alpha * alpha

    # Stall check logic (simplified)
    CL_limit = ac.CL_max_clean if not flaps_down else ac.CL_max_to
    if CL > CL_limit:
        CL = CL_limit * 0.8 # Lift drops in stall

    # CD Calculation
    CD0 = ac.CD0_clean
    if flaps_down: CD0 += ac.CD0_flaps
    if gear_down:  CD0 += ac.CD0_gear

    CD = CD0 + ac.K * CL**2

    # --- Forces ---
    q_dyn = 0.5 * ac.rho * V**2
    Lift = CL * q_dyn * ac.S
    Drag = CD * q_dyn * ac.S
    Thrust = ac.Thrust_max * thrust_level

    # Ground Friction
    mu = 0.02 # Rolling resistance
    Weight = ac.W

    if on_ground:
        # Ground Run Dynamics
        N = max(0, Weight - Lift)
        Friction = mu * N

        V_dot = (Thrust - Drag - Friction) / ac.m
        gamma_dot = 0 # Constrained to ground
        h_dot = 0
        x_dot = V

        # Check for liftoff
        if Lift > Weight:
            # Force liftoff in next step logic effectively handled by alpha logic above
            pass

    else:
        # Flight Dynamics
        # m V_dot = T cos(alpha) - D - W sin(gamma)
        # m V gamma_dot = T sin(alpha) + L - W cos(gamma)

        V_dot = (Thrust * np.cos(alpha) - Drag - Weight * np.sin(gamma)) / ac.m

        # Avoid singularity at V=0
        if V > 1.0:
            gamma_dot = (Thrust * np.sin(alpha) + Lift - Weight * np.cos(gamma)) / (ac.m * V)
        else:
            gamma_dot = 0

        h_dot = V * np.sin(gamma)
        x_dot = V * np.cos(gamma)

        # Ground collision check
        if h <= 0 and h_dot < 0:
            h = 0
            h_dot = 0
            gamma = 0

    return [V_dot, gamma_dot, h_dot, x_dot]

# ==========================================
# 4. EXECUTION AND PLOTTING
# ==========================================

def run_simulation():
    b787 = B787()

    # Simulation scenarios
    scenarios = ['flaps'] # Modified to only simulate 'flaps'
    colors = {'nominal': 'green', 'flaps': 'orange', 'engine': 'red'}
    labels = {
        'nominal': 'Nominal Takeoff',
        'flaps': 'Flaps Retracted Error',
        'engine': 'Engine Fuel Cutoff'
    }

    # Time settings
    t_span = (0, 120) # 2 minutes
    t_eval = np.linspace(0, 120, 1200)

    # Initial State: [V=0, gamma=0, h=0, x=0]
    y0 = [0.01, 0, 0, 0] # V cannot be exactly 0 for gamma_dot div

    results = {}

    for sc in scenarios:
        print(f"Simulating scenario: {sc}...")
        sol = solve_ivp(
            flight_dynamics,
            t_span,
            y0,
            args=(b787, sc),
            t_eval=t_eval,
            method='RK45'
        )
        results[sc] = sol

    # --- Plotting ---
    fig, axs = plt.subplots(2, 1, figsize=(12, 10))

    # Plot 1: Altitude vs Distance (Trajectory)
    ax = axs[0]
    ax.set_title(f"Trajectory Comparison (B787-8 @ {b787.W:.0f} lbs)")
    ax.set_xlabel("Distance [ft]")
    ax.set_ylabel("Altitude [ft]")
    ax.grid(True)

    for sc in scenarios:
        sol = results[sc]
        # Convert x to nm for readability if needed, staying in ft for now
        ax.plot(sol.y[3], sol.y[2], label=f"Sim: {labels[sc]}", color=colors[sc], linewidth=2)

        # Try to load actual data if files exist (based on Untitled3.ipynb logic)
        # Filenames inferred from project description
        fname_map = {'nominal': 'Nominal Takeoff.txt', 'flaps': 'Flap.txt', 'engine': 'Engine Shutdown.txt'}

        actual_data = load_trajectory_data(fname_map[sc])
        if actual_data:
             try:
                 # Corrected column names based on available keys
                 ax.scatter(actual_data['dist___ft'][::50], actual_data['cgftmsl'][::50],
                            marker='x', color=colors[sc], alpha=0.5, label=f"Actual: {labels[sc]}")
             except KeyError as ke:
                 print(f"Could not plot actual data for {sc}: Column name '{ke}' not found. Available keys: {list(actual_data.keys())}")
             except Exception as e:
                 print(f"An unexpected error occurred while plotting actual data for {sc}: {e}")

    ax.legend()
    ax.set_ylim(0, 3000)
    ax.set_xlim(0, 40000)

    # Plot 2: Velocity vs Time
    ax = axs[1]
    ax.set_title("Velocity Profile")
    ax.set_xlabel("Time [s]")
    ax.set_ylabel("Calibrated Airspeed [kts]")
    ax.grid(True)

    for sc in scenarios:
        sol = results[sc]
        # Convert ft/s to knots
        vkts = sol.y[0] / 1.688
        ax.plot(sol.t, vkts, label=f"Sim: {labels[sc]}", color=colors[sc], linewidth=2)

    ax.legend()

    plt.tight_layout()
    plt.show()

    # --- Analysis Output ---
    print("\n--- Simulation Analysis ---")
    # Only provide analysis for 'flaps' since only that scenario is run
    if 'flaps' in results:
        flap_h = results['flaps'].y[2][-1]
        print(f"Final Altitude (Flaps Error): {flap_h:.0f} ft")
        print("\nNote: Adjust 'failure_trigger_h' or aerodynamic coefficients to tune the failure severity.")
    else:
        print("No flaps scenario results to analyze.")


if __name__ == "__main__":
    run_simulation()
```
